# Clustering - Univariate, Bivariate, Multivariate

```python
clustering1 = KMeans(n_clusters=3)
```

```python
clustering1.fit(df[['Annual Income (k$)']])
```

```python
clustering1.labels_
```

```python
df['Income Cluster'] = clustering1.labels_
```

```python
df.head()
```

```python
df['Income Cluster'].value_counts()
```

```python
clustering1.inertia_ #this will give the squared sum of distance of a data point from its assigned center
```

```python
inertia_scores=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df[['Annual Income (k$)']])
    inertia_scores.append(kmeans.inertia_)
```

```python
inertia_scores
```

```python
plt.plot(range(1,11),inertia_scores) #this is the elbow method to determine the number of clusters to be used
```

```python
#now go above and set the number of clusters to 3
```

```python
df.columns
```

```python
df.groupby('Income Cluster')['Age','Annual Income (k$)','Spending Score (1-100)'].mean()
```

```python
#bivariate clustering
```

```python
clustering2 = KMeans(n_clusters=5)
clustering2.fit(df[['Age','Annual Income (k$)','Spending Score (1-100)']])
df['Spending and Income Cluster']=clustering2.labels_
df.head()
```

```python
inertia_scores2=[]
for i in range(1,11):
    kmeans2=KMeans(n_clusters=i)
    kmeans2.fit(df[['Age','Annual Income (k$)','Spending Score (1-100)']])
    inertia_scores2.append(kmeans2.inertia_)
plt.plot(range(1,11),inertia_scores2)
```

```python
#by the elbow method, we can see that the appropriate number of clusters should be 5
```
